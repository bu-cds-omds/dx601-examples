#### Chart: Convex vs Non-Convex Functions

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the functions
def convex_func(x):
    return x**2

def non_convex_func(x):
    return x**4 - 4*x**2 + 2*x

x = np.linspace(-2.5, 2.5, 400)
y1 = convex_func(x)
y2 = non_convex_func(x)

# Points for interpolation
x_pts = np.array([-1, 1])
y1_pts = convex_func(x_pts)
y2_pts = non_convex_func(x_pts)

# Calculate the linear interpolation (secant line) values across the range
# using the formula for a line between (-1, f(-1)) and (1, f(1))
def get_interp_line(x, x_pts, y_pts):
    return np.interp(x, x_pts, y_pts)

y1_interp = get_interp_line(x, x_pts, y1_pts)
y2_interp = get_interp_line(x, x_pts, y2_pts)

# Plot 1: Convex
ax1 = plt.gca()
ax1.plot(x, y1, color='blue', lw=2)
ax1.plot(x_pts, y1_pts, 'ko--', label='Interpolation')
# Shade between the curve and the interpolation line
ax1.fill_between(x, y1, y1_interp, where=(x >= -1) & (x <= 1), color='blue', alpha=0.2)
ax1.set_title("Convex Function")
ax1.text(0, 5, r"$f(x) = x^2$", fontsize=14, ha='center', bbox=dict(facecolor='white', alpha=0.5))
ax1.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()

# Plot 2: Non-Convex
ax2 = plt.gca()
ax2.plot(x, y2, color='red', lw=2)
ax2.plot(x_pts, y2_pts, 'ko--', label='Interpolation')
# Shade between the curve and the interpolation line
ax2.fill_between(x, y2, y2_interp, where=(x >= -1) & (x <= 1), color='red', alpha=0.2)
ax2.set_title("Non-Convex Function")
ax2.text(0, 15, r"$f(x) = x^4 - 4x^2 + 2x$", fontsize=14, ha='center', bbox=dict(facecolor='white', alpha=0.5))
ax2.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```
