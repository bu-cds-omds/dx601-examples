```python
import numpy as np

A = np.array([[0.5, 1.0], [1.0, 0.5]])
A
```

The singular value decomposition can be computed with the `np.linalg.svd` function from NumPy.
It returns a NumPy array for each of the components of the decomposition $\mathbf{U} \mathbf{\Sigma} \mathbf{V}^\top$.

```python
U, s, Vt = np.linalg.svd(A)
```

```python
U
```

The $\mathbf{\Sigma}$ matrix is returned as a vector `s` with just the diagonal entries.

```python
s
```

NumPy's `np.diag` function will return the full diagonal matrix $\mathbf{\Sigma}$.

```python
np.diag(s)
```

The transposed $\mathbf{V}$ matrix is returned for convenience, but the original $\mathbf{V}$ is easily available.

```python
Vt
```

```python
Vt.T
```
