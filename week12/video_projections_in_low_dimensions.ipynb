## Projecting onto Vectors

When we previously talked about projecting onto vectors, one of the formulas for projecting onto a unit vector $\mathbf{b}$ was

\begin{array}{rcl}
\mathrm{proj}_{\mathbf{b}}(\mathbf{a}) & = & (\mathrm{cos}~\theta~\| \mathbf{a} \|)~\mathbf{b} \\
\end{array}

If you repeat the projection,

\begin{array}{rcl}
\mathrm{proj}_{\mathbf{b}}(\mathrm{proj}_{\mathbf{b}}(\mathbf{a})) & = & (\mathrm{cos}~0~\|\mathrm{proj}_{\mathbf{b}}(\mathbf{a}) \|)~\mathbf{b} \\
& = & \|\mathrm{proj}_{\mathbf{b}}(\mathbf{a}) \|~\mathbf{b} \\
& = & (\mathrm{cos}~\theta~\| \mathbf{a} \|)~\mathbf{b} \\
\end{array}

## Projecting onto Vectors with Matrices

Since $\mathbf{b}$ is a unit vector, we can also calculate

\begin{array}{rcl}
\mathrm{proj}_{\mathbf{b}}(\mathbf{a}) & = & (\mathrm{cos}~\theta~\| \mathbf{a} \|)~\mathbf{b} \\
& = & (\mathrm{cos}~\theta~\| \mathbf{a} \| \| \mathbf{b} \|)~\mathbf{b} \\
& = & (\mathbf{a} \cdot \mathbf{b})~\mathbf{b} \\
& = & (\mathbf{a}^T \mathbf{b}) \mathbf{b} \\
& = & ((\mathbf{a}^T \mathbf{b}) \mathbf{b}^T)^T \\
\end{array}

## Projecting onto Vectors with Matrices

We can cleanup those transposes with the matrix multiplication identity $(\mathbf{A} \mathbf{B})^T = \mathbf{B}^T \mathbf{A}^T$.

\begin{array}{rcl}
\mathrm{proj}_{\mathbf{b}}(\mathbf{a}) & = & ((\mathbf{a}^T \mathbf{b}) \mathbf{b}^T)^T \\
& = & \mathbf{b}~(\mathbf{a}^T \mathbf{b})^T \\
& = & \mathbf{b}~(\mathbf{b}^T~\mathbf{a}) \\
& = & (\mathbf{b}~\mathbf{b}^T)~\mathbf{a} \\
\end{array}

So $\mathbf{b}~\mathbf{b}^T$ is a projection matrix that projects onto the vector $\mathbf{b}$.

## Projecting onto a Plane

Suppose we want to project from 3 dimensions onto a plane?
* The plane must go through the origin.
* Pick any unit vector $\mathbf{b}$ that is orthogonal to the plane.
* Multiplying with $\mathbf{b}~\mathbf{b}^T$ projects onto $\mathbf{b}$.
* Multiplying with $I - \mathbf{b}~\mathbf{b}^T$ projects onto the plane.

## Projecting to the Closest Point in 2D


```python

import matplotlib.pyplot as plt
import numpy as np

P = np.asarray([[0.5, 0.5], [0.5, 0.5]])

plt.figure(figsize=(4, 4))

plt.plot([-5, 5], [-5, 5], label="projected outputs")

def project(x, y):
    projected = P @ np.asarray([[x], [y]])
    x2 = projected[0,0]
    y2 = projected[1, 0]

    plt.arrow(x, y, x2 - x, y2 - y, head_width=0.2, length_includes_head=True)

project(-4, 4)
project(-2.2, 3.2)
project(-1.1, 3.9)
project(0, 4)
project(2.1, 3.6)
project(3.5, -3.3)
project(4.6, -2.5)

plt.legend()
plt.show()

print(P)

None
```
