# Video: Plotting Histograms

This video shows how to plot histograms of data in Python using the library Matplotlib.

```python
import matplotlib.pyplot as plt
```

```python
data = [0, 0, 1, 0, 2, 3, 5]
```

```python
_ = plt.hist(data)
```

```python
_ = plt.hist(data, bins=6, range=(0,6))
```

```python
import numpy as np
```

```python
data2 = np.random.normal(size=10000)
```

```python
data2
```

```python
_ = plt.hist(data2, bins=10)
```

```python
_ = plt.hist(data2, bins=100)
```
