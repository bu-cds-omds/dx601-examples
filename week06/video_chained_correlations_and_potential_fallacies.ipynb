## Chained Correlations and Potential Fallacies

* A is correlated with B
* B is correlated with C
* Is A correlated with C?

* Diets high in cholesterol are correlated with higher blood cholesterol.
* Higher cholesterol is correlated with clogged arteries.
* Clogged arteries are correlated with cardiac risk.

Script:

* That recommendation has received a lot of criticism over the years for not being supported by the study, but at the time, it seemed like it was confirming a chain of already known steps.
* Dietary cholesterol was correlated with blood cholesterol which was correlated with clogged arteries which was correlated with various cardiac risk such as heart attacks.
* In fact, usually, those correlations were stated as causations.
* High dietary cholesterol caused high blood cholesterol which caused clogged arties which caused cardiac risk.
* So therefore, high dietary cholesterol causes cardiac risk?
* Maybe?

## Cholesterol and Cardiac Risk Spoiler

* Mostly yes, but the popular media gets the details wrong.

Script:
* Before going into the details, I'd like to say ahead of time that this is not going to be an anti science or anti medicine rant.
* The research has mostly supported this chain, but not nearly as strongly as the low fat policy and popular press would suggest.
* In particular, most of the studies only support claims for correlation, not causation, because running long term studies controlling what a lot people eat is difficult.
* So this is an example I tend to pick on for reading too much into correlations and causations.
* I am not picking on the science per se, but the hype and jumping to conclusions that followed the science.

## Cholesterol and Cardiac Risk Concerns

* Diets high in cholesterol are correlated with higher blood cholesterol.
  * Original study based on rabbits.
  * Later human studies say dietary cholesterol only explains 1% of variation???
* Higher cholesterol is correlated with clogged arteries.
  * Same rabbit studies.
* Clogged arteries are correlated with cardiac risk.
  * I believe this is well established for humans.
  * See coronary artery bypass surgery.

Script:
* As I said before, the steps in this chain all had evidence to support them, but some of it was dicey to include in a multi-step inference for humans.
* You can take this slide as a snapshot of my personal knowledge between 2010 and 2015, bearing in mind that my PhD is in computer science, not medicine.
* The strongest evidence connecting dietary and blood cholesterol was from a study about rabbits.
* They force fed rabbits a large fraction of their body mass in cholesterol, and their blood cholesterol was sky high and their arteries were totally clogged.
* Kind of makes sense?
* But rabbits do not normally have cholesterol in their diet; that comes from eating meat.
* So rabbits probably react to eating cholesterol differently than humans who often do eat meat.
* Clogged arteries are pretty well known to be connected to cardiac risk; coronary artery bypass surgery is working around these clogs.
* So, yes evidence, but absolutely not a solid 100% sure path to eating fat gives heart attacks.
* But science does make progress with time, and there have been more studies since then.

## A Standard Synthetic Example

Suppose $X$ and $Y$ are independent random variables chosen uniformly between zero and one, then the correlation matrix of $X$, $X+Y$, and $Y$ is

| $\rho$ | X | X+Y | Y |
|:---:|:---:|:---:|:---:|
| **X** | 1 | 0.7 | 0 |
| **X+Y** | 0.7 | 1 | 0.7 |
| **Y** | 0 | 0.7 | 1 |

$X+Y$ has approximately 0.7 correlation with both $X$ and $Y$, but $X$ and $Y$ have zero correlation.

Script:
* This is the standard counterexample to chains of correlated variables that I've seen in textbooks.
* If you have two independent uniform random variables X and Y, then their correlation is zero because they are independent.
* But both X and Y are about 70% correlated with the sum X plus Y.
* So a chain of 70% correlations can become zero correlation in just two steps.
* Is that too artificial?
* Here is another example that while still synthetic, shows how drift can accumulate over the chain.

```python
#@title ##### Slide: Two Highly Correlated Functions

import math
import matplotlib.pyplot as plt
import numpy as np

def get_data(offset):
    return [math.sin((x - offset * 25) / 50) for x in range(1000)]

def plot_offset(offset):
    plt.plot(get_data(offset))

offset_0 = 0
offset_1 = 1

plot_offset(offset_0)
plot_offset(offset_1)

correlation = np.corrcoef(get_data(offset_0), get_data(offset_1))[0][1]
_ = plt.title(f"Correlation {correlation:.2f}")
```

Script:
* These two functions are identical except one has been shifted a bit to the right.
* They have about 87% correlation.
* That's pretty high.
* If we keep repeating the same shift, consecutive pairs will keep that correlation.

```python
#@title ##### Slide: A Long Chain of Highly Correlated Functions

for o in range(7):
    plot_offset(o)
```

Script:
* But after six shifts, they are completely out of phase.

```python
#@title ##### Slide: Ends of A Long Chain of Highly Correlated Functions

offset_2 = 6
plot_offset(offset_0)
plot_offset(offset_2)
correlation = np.corrcoef(get_data(offset_0), get_data(offset_2))[0][1]
_ = plt.title(f"Correlation {correlation:.2f}")
```

Script:
* Almost a perfect negative one correlation.
* Even though the correlation was so high at each step.

## Chained Correlations Summary

* Correlations can drop very quickly over multiple steps in a chain.
* Real world correlations tend to be much lower, and have other factors mixed in.
* Be very careful before assuming they hold.


Script:
* Wrapping up, be careful assuming correlations hold along a chain.
* Often, they do not hold at all, or the magnitude of the correlation drops quite a bit.
