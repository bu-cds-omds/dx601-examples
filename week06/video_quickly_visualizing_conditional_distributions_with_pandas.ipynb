# Video: Quickly Visualizing Conditional Distributions with Pandas

This video shows how to quickly visualize conditional distributions with pandas using examples from the abalone data set.

## Quickly Visualizing Conditional Distributions with Pandas

* Focus on visualizations
* Skip probabilities.
* Use pandas visualization directly.

```python
import pandas as pd
```

```python
abalone = pd.read_csv("https://raw.githubusercontent.com/bu-cds-omds/dx601-examples/main/data/abalone.tsv", sep="\t")
```

```python
_ = abalone.hist("Rings")
```

```python
_ = abalone.hist("Rings", density=True)
```

```python
_ = abalone.hist("Rings", bins=100)
```

```python
_ = abalone.query("Sex == 'M'").hist("Rings", bins=100)
```

```python
_ = abalone.query("Sex == 'F'").hist("Rings", bins=100)
```

```python
_ = abalone.query("Sex == 'I'").hist("Rings", bins=100)
```

```python
_ = abalone.hist("Diameter", bins=100)
```

```python
_ = abalone.query("Diameter <= 0.3").hist("Rings")
```

```python
#@title ##### Slide: Plot Conditional Distributions to Check Your Claims

import matplotlib.pyplot as plt

abalone.query("Diameter <= 0.3").hist("Rings")
plt.annotate("My Expected Mode", xy=(5,125), xytext=(1,150), arrowprops=dict(facecolor='black', shrink=0.05))
plt.annotate("The Actual Mode", xy=(8.5,200), xytext=(10,200), arrowprops=dict(facecolor='black', shrink=0.05))
None
```
