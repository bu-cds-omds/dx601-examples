```python
import matplotlib.pyplot as plt
import numpy as np

def plot_transformation(A, title):
    figure = plt.figure(figsize=(3, 5))
    ax = figure.add_subplot(111)

    # Define the unit square vertices: (0,0), (1,0), (1,1), (0,1)
    square = np.array([
        [0, 1, 1, 0, 0],
        [0, 0, 1, 1, 0]
    ])

    # Transform the square
    transformed_square = A @ square

    # Calculate determinant
    det = np.linalg.det(A)

    # Plot original square (dashed)
    ax.plot(square[0, :], square[1, :], 'k--', alpha=0.3, label='Original')

    # Plot transformed square
    ax.fill(transformed_square[0, :], transformed_square[1, :], color='skyblue', alpha=0.5, label='Transformed')
    ax.plot(transformed_square[0, :], transformed_square[1, :], color='blue', lw=2)

    # Formatting
    ax.axhline(0, color='black', lw=1)
    ax.axvline(0, color='black', lw=1)
    ax.set_title(f"{title}\nDet = {det:.2f}", fontsize=10)
    ax.set_aspect('equal')
    ax.grid(True, linestyle=':', alpha=0.6)

    # Display Matrix A in plain text to avoid LaTeX parsing errors
    matrix_text = f"A = [[{A[0,0]:.1f}, {A[0,1]:.1f}],\n     [{A[1,0]:.1f}, {A[1,1]:.1f}]]"
    ax.text(0.5, -0.3, matrix_text, ha='center', va='top', transform=ax.transAxes, fontsize=9, family='monospace')
    ax.set_xlim(-1, 2.5)
    ax.set_ylim(-1, 2.5)

    figure.show()


# Define Matrices
A1 = np.eye(2)                                      # Identity
A2 = np.array([[1, 1], [0, 1]])                      # Horizontal Skew
# Rotate 45 deg and stretch by sqrt(2) along x=y
theta = np.pi/4
rot = np.array([[np.cos(theta), -np.sin(theta)], [np.sin(theta), np.cos(theta)]])
stretch = np.array([[2, 0], [0, 1]])
A3 = rot @ stretch @ rot.T
# Project onto y=x line
A4 = np.array([[0.5, 0.5], [0.5, 0.5]])

# Create 2x2 grid and flatten axes for easy indexing
# fig, axes = plt.subplots(2, 2, figsize=(12, 12))
# ax_flat = axes.flatten()

plot_transformation(A1, "Identity Matrix")
plot_transformation(A2, "Horizontal Skew")
plot_transformation(A3, "Rotate & Stretch")
plot_transformation(A4, "Project onto y=x")

# for ax in ax_flat:
#     ax.set_xlim(-1, 2.5)
#     ax.set_ylim(-1, 2.5)

plt.tight_layout()
plt.show()
```
