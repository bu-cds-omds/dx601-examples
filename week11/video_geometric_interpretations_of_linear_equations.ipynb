# Video: Geometric Interpretations of Linear Equations

This video shows the common geometric interpretations of linear equations.




## Geometric Interpretation of Linear Equations by Number of Variables

* 0 variables: all or nothing
* 1 variable: fixed value
* 2 variables: line
* 3 variables: plane
* 4+ variables: hyperplane


Script:
* The default geometric interpretation of a linear equation depends on the number of non-zero coefficients.
* Zero variables is an all or nothing condition, more like a boolean.
* One variable is actually a constant.
* Two variables determine a line.
* Three variables determine a plane.
* We can draw all of these so far.
* Four or more variables define a hyperplane, and drawing them becomes difficult.


## Why Ignore Zero Coefficients?

\begin{array}{rcl}
x + 2y & = & 3 \\
0a + 0b + 0c + 0d + x + 2y + 0z & = & 3 \\
\end{array}

Script:
* Variables with zero coefficients do not change the meaning of the equation, so we ignore them unless we have context that makes them worth mentioning.
* An example of that would be another equation in the same system of linear equations mentioning them.

## 0 Variables is All or Nothing

If all coefficients are zero, then the generic equation

\begin{array}{rcl}
\sum_{j=0}^{n-1} a_j x_j & = & b \\
\end{array}

reduces to

\begin{array}{rcl}
0 & = & b \\
\end{array}

Is $b$ zero?


Script:
* If all the variable coefficients are zero, so none of the variables are actually used, then our linear equation simplifies to comparing zero and the constant on the right b.
* So if b is zero, then the equation is always true and all variable assignments are solutions.
* If b is not zero, then the equation is never true and no variable assignments are solutions.
* So the zero variable case is the entire universe of assignments or nothing at all.

## 1 Variable is a Actually a Constant

The one variable case looks like

\begin{array}{rcl}
a_i x_i & = & b \\
\end{array}

So this case is actually fixing the variable at specific value, so it is more like a constant.

\begin{array}{rcl}
x_i & = & \frac{b}{a_i} \\
\end{array}



Script:
* If only one variable has a non-zero coefficient, the linear equation simplifies, and we can easily derive the value of that variable.
* The geometric interpretation is simple.
* We draw a number line and show the single point on the number line with the only allowed value for $x$.

## 1 Variable on a Number Line


```python
import matplotlib.pyplot as plt

plt.axis('off')
plt.gca().axhline(y=0, color="grey")
plt.gca().set_xlim(-2, 12)
plt.gca().set_ylim(-20, 20)
plt.plot([6], [0], label="2x = 12", linestyle="none", marker="o")
plt.legend()

for x in range(-1, 12):
    plt.vlines(x=x, ymin=-1, ymax=1, colors="grey")
    plt.text(x=x, y=-3, s=str(x), horizontalalignment="center")

None
```

Script:
* This visualization is not particularly exciting by itself, but is replicated for each axis when we look at higher dimensions.

## 2 Variables Determine a Line


```python
import matplotlib.pyplot as plt
import numpy as np

xs = np.linspace(-2, 8)

plt.gca().axhline(y=0, color="grey")
plt.gca().axvline(x=0, color="grey")

plt.plot(xs, [(12 - 2 * x) / 3 for x in xs], label="2x + 3y = 12")
plt.plot([0, 6], [4, 0], color="orange", linestyle="none", marker="o",  label="intercepts")

arrowprops = {
    "facecolor": "black",
    "headlength": 5,
    "headwidth": 3,
    "shrink": 0.1,
    "width": 0.5
    }

plt.annotate("x-intercept",
             arrowprops=arrowprops,
             xy=(6,0),
             xytext=(6.25,1))

plt.annotate("y-intercept",
             arrowprops=arrowprops,
             xy=(0,4),
             xytext=(1, 4.5))

plt.legend()

None
```

Script:
* Here is a chart of solutions to the equation 2x plus 3y equals 12.
* We can easily calculate the x and y intercepts where this line meets the x and y axes respectively.
* Then given those two points, we can draw the unique line through both of them.

## 2 Variables, Calculating the X and Y Intercepts

* Given:
  * $a_0 x + a_1 y = b$
* The $x$ intercept is the $x$ value when $y = 0$.
  * $a_0 x = b$
  * $x = \frac{b}{a_0}$
* The $y$ intercept is the $y$ value when $x = 0$.
  * $a_1 y = b$
  * $y = \frac{b}{a_1}$


Script:
* How do we calculate those intercepts?
* The x intercept is where the line cross, or intercepts, the x axis, so y = 0.
* If you set y equal to zero in the equation for the line, then the equation simplifies to $a_0 x = b$, which we then solve exactly like before in the one variable case.
* So $x = \frac{b}{a_0}$.
* Similarly, the y intercept is where the line crosses the y axis, so $x = 0$, and $y = \frac{b}{a_1}$.

## 2 Variables Interpolate X and Y Intercepts


```python
import matplotlib.pyplot as plt
import numpy as np

xs = np.linspace(-2, 8)

plt.gca().axhline(y=0, color="grey")
plt.gca().axvline(x=0, color="grey")

plt.plot(xs, [(12 - 2 * x) / 3 for x in xs], label="2x + 3y = 12")
plt.plot([0, 6], [4, 0], color="orange", linestyle="none", marker="o",  label="intercepts")

arrowprops = {
    "facecolor": "black",
    "headlength": 5,
    "headwidth": 3,
    "shrink": 0.1,
    "width": 0.5
    }

plt.annotate("x-intercept at\nx = 12/2",
             arrowprops=arrowprops,
             xy=(6,0),
             xytext=(6.25,1))

plt.annotate("y-intercept at\ny = 12/3",
             arrowprops=arrowprops,
             xy=(0,4),
             xytext=(1, 4.5))

plt.legend()

None
```

Script:
* Once we know the x and y intercepts, we can draw a line between them.
* We can generalize this process to higher dimensions.

## Generalized Axis Intercepts

The 1 variable case is
* $x_i = \frac{b}{a_i}$

With more variables, the $x_i$ intercept with the $x_i$ axis is at
* $x_i = \frac{b}{a_i}$, and $x_j = 0$ for all $j \neq i$

So draw those intercepts, and interpolate between them.

Script:
* Here is the generalization of that one variable calculation.
* As with the two dimensional case, all the other variables are set to zero and their terms drop out.
* That leaves the equation as $a_i x_i = b$, and we divide by $a_i$ to get $x_i = \frac{b}{a_i}$.
* And we know that division always works because we are explicitly focusing on the variables with non-zero coefficients.

## 4+ Variables Determine a Hyperplane

* Given:
  * $2 x_0 + 3 x_1 + 4 x_2 + 6 x_3 = 12$.
* Then the intercepts are at
  * $x_0 = \frac{12}{2} = 6$
  * $x_1 = \frac{12}{3} = 4$
  * $x_2 = \frac{12}{4} = 3$
  * $x_3 = \frac{12}{6} = 2$.
* How to draw four axes?



Script:
* Increasing past three dimensions becomes harder to visualize.
* We are used to seeing three dimensional scenes in two dimensions.
* Every photo is an example of this.
* But visualizing four or more dimensions is tricky in 2 dimensions.
* In my opinion, they either look artificially regular, or give misleading intuitions about the geometry.


## 2D Interpretations With 1 Variable


```python
import matplotlib.pyplot as plt
import numpy as np

plt.gca().set_xlim(-5, 5)

plt.gca().axhline(y=0, color="grey")
plt.gca().axvline(x=0, color="grey")

plt.plot([2, 2], [-5, 5], color="blue", label="2x + 0y = 4")

arrowprops = {
    "facecolor": "black",
    "headlength": 5,
    "headwidth": 3,
    "shrink": 0.1,
    "width": 0.5
    }

plt.annotate("x-intercept at\nx = 4/2",
             arrowprops=arrowprops,
             xy=(2,0),
             xytext=(2.25,1))

plt.legend()

None
```

Script:
* Previously, we just counted the non-zero coefficients.
* A zero coefficient for a variable means that variable does not matter.
* You can change it arbitrarily, and it does not matter if the linear equation multiplies that value by zero.
* If we look at this equation, 2x + 0y = 4, the y coefficient is zero and the equation simplifies to 2x = 4.
* So we solve that and get x equals 2.
* If we bring back the y to draw a line, the y still doesn't matter, so our line is all values where x equals 2, and y has any value.
* So that gives us the vertical line at x equals 2.
* We cannot use the intercept interpolation because there is no longer an intercept.
* If we tried to calculate the intercept like normal, we would divide 4 by zero which does not work.
* Instead, the zero coefficient means that the surface we are describing is parallel to that axis and never meets it.

## Interpretation of Systems of Linear Equations


```python
import matplotlib.pyplot as plt
import numpy as np

xs = np.linspace(-2, 8)

plt.gca().axhline(y=0, color="grey")
plt.gca().axvline(x=0, color="grey")

plt.plot(xs, [(12 - 2 * x) / 3 for x in xs], label="2x + 3y = 12")
plt.plot(xs, [(10 - 5 * x) / 2 for x in xs], label="5x + 2y = 10")

y_solution = 40 / 11
x_solution = (12 - 3 * y_solution) / 2
plt.plot([x_solution], [y_solution], label="solution", linestyle="none", marker="o")

plt.legend()

None
```

Script:
* When we talk about a system of linear equations, we are talking about variable assignments that are solutions to all of the linear equations at once.
* Geometrically, that means the intersection of the solutions to the individual equations.
* In this figure, the solution is where the lines cross.
* If there are more than two equations, we mean the intersections of all the equations at once.

## No Intersection = Inconsistent System


```python
import matplotlib.pyplot as plt
import numpy as np

xs = np.linspace(-2, 8)

plt.subplot(1, 2, 1)

plt.gca().axhline(y=0, color="grey")
plt.gca().axvline(x=0, color="grey")

plt.plot(xs, [(12 - 2 * x) / 3 for x in xs], label="2x + 3y = 12")
plt.plot(xs, [(10 - 2 * x) / 3 for x in xs], label="2x + 3y = 10")

plt.legend()

plt.subplot(1, 2, 2)

plt.gca().axhline(y=0, color="grey")
plt.gca().axvline(x=0, color="grey")

plt.plot(xs, [(12 - 2 * x) / 3 for x in xs], label="2x + 3y = 12")
plt.plot(xs, [(10 - 5 * x) / 2 for x in xs], label="5x + 2y = 10")
plt.plot(xs, [(5 + 3 * x) / 4 for x in xs], label="-3x + 4y = 5")

plt.legend()

None
```

Script:
* If there are no points where all the equations meet, there is no solution to the system.
* This may mean that some of the equations ever intersect, like the chart on the left, where the two lines are parallel.
* Or that some of the equations intersect, but not all of them at a single point, like the chart on the right.

## Homogenous Systems of Linear Equations


```python
import matplotlib.pyplot as plt
import numpy as np

xs = np.linspace(-8, 8)

plt.gca().axhline(y=0, color="grey")
plt.gca().axvline(x=0, color="grey")

for c in range(10):
    plt.plot(xs, [0 - c * x for x in xs], label=f"{c}x + y = 0")

plt.legend()

None
```

Script:
* When I described how to draw the solutions based on the axis intercepts, I skipped the one case where this scheme does not work.
* If the right hand constant b is zero, then intercepts collapse into one point at the origin.
* Otherwise, all the intercepts are distinct points.
* In two dimensions, that looks like this picture.
* The classification into points, lines, and planes still works, just the axis interpolation drawing does not work.
* An interesting thing about this case is that setting all variables to zero always works.
* You always end up with zeros on both sides.
* This means that a homogenous system of linear equations always has a solution at the origin.
* No matter how many equations you add, as long as the system remains homogenous, the origin setting all variables to zero is always a solution.
