This video illustrates how linear dependencies and collinearities can affect the number of solutions to a system of linear equations.

## Intuitions about Linear Dependencies and Collinearities

* Linear dependencies are redundancies.
* Collinearities are a special case.

## Linear Dependencies in Systems of Linear Equations?

* What does it mean if a system of linear equations has a linear dependency?
  1. One of the equations can be expressed as a linear combination of the other equations.
  2. That equation is redundant.
  3. That equation can be removed without changing the solutions.


## An Example Linear Dependency

\begin{array}{rcl}
3x + 6y + 7z & = & 0 \\
5x + 3y + 0z & = & 3 \\
8x + 9y + 7z & = & 3 \\
\end{array}

## A 2D Example Linear Dependency


```python

import matplotlib.pyplot as plt
import numpy as np

xs = np.linspace(-2, 8)

plt.gca().axhline(y=0, color="grey")
plt.gca().axvline(x=0, color="grey")

plt.plot(xs, [(12 - 2 * x) / 3 for x in xs], label="2x + 3y = 12")
plt.plot(xs, [(10 - 5 * x) / 2 for x in xs], label="5x + 2y = 10")
plt.plot(xs, [(22 - 7 * x) / 5 for x in xs], label="7x + 5y = 22")

plt.legend()

None
```

## More Linear Dependencies


```python

import matplotlib.pyplot as plt
import numpy as np

xs = np.linspace(-2, 8)

plt.gca().axhline(y=0, color="grey")
plt.gca().axvline(x=0, color="grey")

for alpha in np.linspace(0, 1, num=11):
    plt.plot(xs, [alpha * (12 - 2 * x) / 3 + (1 - alpha) * (10 - 5 * x) / 2 for x in xs],
             label=f"alpha = {alpha:.2f}")

plt.legend()

plt.annotate(text="alpha=0.00 => 2.0x + 3.0y = 12\nalpha=0.50 => 3.5x + 2.5y = 11\nalpha=1.00 => 5.0x + 2.0y = 10",
             xy = (3, 5))

None
```

## Collinearity, A Special Case of Linear Dependency

\begin{array}{rcl}
2x + 3y & = & 12 \\
(2 \alpha) x + (3 \alpha) y & = & 12 \alpha \\
\end{array}

## A 2D Example of Collinearity


```python

import matplotlib.pyplot as plt
import numpy as np

xs = np.linspace(-2, 8)

plt.gca().axhline(y=0, color="grey")
plt.gca().axvline(x=0, color="grey")

plt.plot(xs, [(12 - 2 * x) / 3 for x in xs], label="2x + 3y = 12")
plt.plot(xs, [(24 - 4 * x) / 6 for x in xs], label="4x + 6y = 24")

plt.legend()

None
```

## Why Do Collinear Equations Represent the Same Geometry?

Recall the generic linear equation

\begin{array}{rcl}
\sum_{j=0}^{n-1} a_j x_j & = & b \\
\end{array}

Where the $j$th axis intercept is at $x_j = \frac{b}{a_j}$...


## Comparing Intercepts of Collinear Equations

If both sides of the equation are multiplied by $\alpha \neq 0$,

\begin{array}{rcl}
\sum_{j=0}^{n-1} (\alpha a_j) x_j & = & \alpha b \\
\end{array}

and the $j$th axis intercept is at $x_j = \frac{\alpha b}{\alpha a_j} = \frac{b}{a_j}$.
