## Sampling Boolean Random Variables

* Given a probability $p$, how do we sample a Boolean random variable that is true with probability $p$?
* Python has a `random` module.

Script:
* In this video, we will see how to sample a Boolean random variable with a known probability.
* Python has a random number generation module that makes this quite easy.
* This module is just called random.

```python
import random
```

Script:
* The random module has a function in it, also called random, that returns a pseudorandom number between 0 and 1.
* Technically, that number will be greater than or equal to zero, and strictly less than one.
* Let's try it out.

```python
random.random()
```

```python
random.random()
```

Script:
* Let's look at a lot more samples.

```python
import matplotlib.pyplot as plt
plt.plot([random.random() for _ in range(100)])
```

Script:
* Those points are not really connected.
* They are supposed to be separate random samples.
* I will plot a fresh sample without lines.

```python
plt.plot([random.random() for _ in range(100)], ".")
```

Script:
* Now that looks like a bunch of random points.
* Let's return to our original problem.
* How do we sample from a Boolean random variable when we know the probability that it is true?
* We can do this easily by calling random dot random and just comparing to the probability.
* Here's the code.

```python
def sample_boolean(probability):
    return random.random() < probability
```

```python
sample_boolean(0.5)
```

```python
sample_boolean(0.5)
```

Script:
* How does this work?
* Consider this sampling function with probability one half.


```python
plt.plot([random.random() for _ in range(100)], ".")
plt.plot([0, 100], [0.5, 0.5], label="threshold=0.50")
plt.legend()
```

Script:
* If we generate a random nunber uniformly at random between zero and one, it will be between zero and one half just as often as it is between one half and one.
* The phrase uniform at random is key here.
* Those two ranges are the same size, so we should pick from them equally as often.
* So we will get a random number from each side half the time.
* So when we compare our random dot random output, it will be less than one half, half the time on average.
* Let's look at a higher threshold.


```python
plt.plot([random.random() for _ in range(100)], ".")
plt.plot([0, 100], [0.75, 0.75], label="threshold=0.75")
plt.legend()
```

Script:
* If you increase the probability, the range where we return true, the range from zero to our probability, grows proportionally.
* Let's look at a lower probability.


```python
plt.plot([random.random() for _ in range(100)], ".")
plt.plot([0, 100], [0.25, 0.25], label="threshold=0.25")
plt.legend()
```

Script:
* If you decrease the probability, the range where we return true, the range from zero to our probability, shrinks proportionally.
* What if the theshold is one?


```python
plt.plot([random.random() for _ in range(100)], ".")
plt.plot([0, 100], [1.00, 1.00], label="threshold=1.00")
plt.legend()
```

Script:
* If the probability increases to one, then we are comparing a number between zero and one, but strictly lower than one, comparing if it is less than one.
* This is the check being done.

```python
random.random() < 1
```

Script:
* So if the probability is set to one, then that check always returns True.

```python
sample_boolean(1.0)
```

```python
sample_boolean(1.0)
```

Script:
* If we set the probability to zero, then we are comparing to zero.

```python
plt.plot([random.random() for _ in range(100)], ".")
plt.plot([0, 100], [0.00, 0.00], label="threshold=0.00")
plt.legend()
```

Script:
* If we set the threshold to zero, the uniform random numbers will never below the threshold, so the value will always be false.

```python
random.random() < 0
```

Script:
* So that check will always be false.

## Sampling Boolean Random Variables Recap

To sample a Boolean random variable with probability $p$,
* Sample uniform random $[0,1)$ via `random.random()`
* Threshold that value with $p$ to decide whether sample is true or false.


Script:
* That is it.
* Knowing the probability makes the sampling trivial with the random module.
