# Chart: An Illustration of Convexity


```python
import matplotlib.pyplot as plt
import numpy as np

# Define a convex function
def f(x):
    return x**2

# Define points x1 and x2
x1, x2 = -1.5, 1
y1, y2 = f(x1), f(x2)

# Points between x1 and x2
lambdas = np.linspace(0, 1, 100)
x_line = lambdas * x1 + (1 - lambdas) * x2
y_line = lambdas * y1 + (1 - lambdas) * y2

# Generate x values for the curve
x = np.linspace(-2, 2, 400)
y = f(x)

# Plot
plt.figure(figsize=(7, 5))
plt.plot(x, y, label=r'$f(x) = x^2$', linewidth=2)
plt.plot(x_line, f(x_line), 'r--', label='Function between $x_1$ and $x_2$')
plt.plot(x_line, y_line, 'g--', label='Line between $(x_1, f(x_1))$ and $(x_2, f(x_2))$')
plt.scatter([x1, x2], [y1, y2], color='black')
plt.text(x1 - 0.125, y1 - 0.0625, r'$(x_1, f(x_1))$', ha='right')
plt.text(x2 + 0.125, y2 - 0.0625, r'$(x_2, f(x_2))$', ha='left')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Illustration of Convexity')
plt.legend()
plt.show()
```
