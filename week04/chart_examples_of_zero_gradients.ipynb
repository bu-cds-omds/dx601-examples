# Chart: Examples of Zero Gradients


```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(12, 4))

xs = np.linspace(-2, 2, 100)

plt.subplot(1, 3, 1)
plt.axhline(0, color="grey")
plt.plot(xs, xs ** 2 + 3, label=r"$x^2 - 3$")
plt.plot(xs, 2 * xs, label=r"$\frac{d}{dx} \left(x^2 + 3 \right)$")
plt.title("A Local Minima")
plt.legend()
plt.xlim(-2, 2)

plt.subplot(1, 3, 2)
plt.axhline(0, color="grey")
plt.plot(xs, 4 - xs ** 4, label=r"$4 - x^4$")
plt.plot(xs, -4 * xs ** 3, label=r"$\frac{d}{dx} \left(4 - x^4 \right)$")
plt.title("A Local Maxima")
plt.legend()
plt.xlim(-2, 2)

plt.subplot(1, 3, 3)
plt.axhline(0, color="grey")
plt.plot(xs, xs**3, label=r"$x^3$")
plt.plot(xs, 3 * xs ** 2, label=r"$\frac{d}{dx} \left(x^3\right)$")
plt.title("A Saddlepoint")
plt.legend()
plt.xlim(-2, 2)

plt.show();
```
