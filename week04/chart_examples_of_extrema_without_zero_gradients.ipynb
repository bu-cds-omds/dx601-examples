# Chart: Examples of Extrema without Zero Gradients


```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(12, 4))

xs = np.linspace(-2, 2, 100)

plt.subplot(1, 3, 1)
plt.axhline(0, color="grey")
plt.plot([x for x in xs if x < 0], [1 / x for x in  xs if x < 0], label=r"$1 / x$")
plt.plot([x for x in xs if x > 0], [1 / x for x in  xs if x > 0], color=plt.gca().lines[-1].get_color())
plt.title("Both Extrema at a Discontinuity")
plt.legend()
plt.xlim(-2, 2)

plt.subplot(1, 3, 2)
plt.plot(xs, 2 * xs, label=r"$x$ where $x \in [-2, 2]$")
plt.title("Extrema at Parameter Boundaries")
plt.legend()
plt.xlim(-3, 3)

plt.show()
```
