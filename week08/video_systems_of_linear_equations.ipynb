## What are linear equations?

A linear equation with $n$ variables $x_1, \ldots, x_n$ has the form

\begin{array}{rcl}
\sum_{j=0}^{n-1} a_j x_j & = & b \\
\end{array}


Using vectors $a$ and $x$ to represent the coefficients and variables respectively, this can be expressed with a matrix multiplication as

\begin{array}{rcl}
a^T x & = & b \\
\end{array}

## 2D Linear Equations

For two dimensions, it is common to use the variables $x$ and $y$, coefficients $a$ and $b$, and constant $c$.

\begin{array}{rcl}
ax + by & = & c \\
\end{array}

Or one coefficient $m$ for the $x$ variable, and constant $b$.

\begin{array}{rcl}
y & = & mx + b \\
\end{array}

## 2D Line Equation

This form is convenient for representing lines in the plane.

\begin{array}{rcl}
ax + by & = & c \\
\end{array}

* If $a \neq 0$, the $x$ intercept is $\frac{c}{a}$ when $y = 0$.
* If $b \neq 0$, the $y$ intercept is $\frac{c}{b}$ when $x = 0$.
* The cases $a = 0$ and $b = 0$ correspond to horizontal and vertical lines respectively.

Slide: Plot of 2x + y = 3


```python
import matplotlib.pyplot as plt
import numpy as np

xs = np.linspace(-2, 2)
ys = [3 - 2 * x for x in xs]
plt.plot(xs, ys, label="2x + y = 3")

axis = plt.gca()
axis.axhline(y=0, color='k')
axis.axvline(x=0, color='k')

plt.annotate("y-intercept\n(0, 3/1)", (0, 3), (0.125, 3.25))
plt.annotate("x-intercept\n(3/2, 0)", (1.25, 1.00))

plt.legend()

None
```

## What is a System of Linear Equations?

A system of linear equations is a set of linear equations.

* The system of linear equations represents all the variable assignments that satisfy all the equations at once.
* We call finding those assignments solving the system of linear equations.

## Systems of Linear Equations

We can describe systems of linear equations like this -

\begin{array}{rcl}
\forall_i \sum_{j=0}^{n-1} a_{i,j} x_j & = & b_i \\
\end{array}

With matrix multiplication, we can write that as

\begin{array}{rcl}
Ax & = & b \\
\end{array}

with $A$ being a matrix of $a_{i,j}$ values, $x$ a vector of the variables, and $b$ a vector of the constants.


## Solving 2x + y = 3 and -x + 4y = 2


```python
import matplotlib.pyplot as plt
import numpy as np

xs = np.linspace(-2, 2)
plt.plot(xs, [3 - 2 * x for x in xs], label="2x + y = 3")
plt.plot(xs, [0.5 + 0.25 * x for x in xs], label="-x + 4y = 2")

x_solution = 10 / 9
y_solution = 7 / 9
plt.plot([x_solution], [y_solution], "*k", label="solution")

axis = plt.gca()
axis.axhline(y=0, color='k')
axis.axvline(x=0, color='k')

plt.legend()

None
```
