```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Create a structured grid for lines
n_lines = 20
u = np.linspace(0, 2 * np.pi, n_lines)
v = np.linspace(0, np.pi, n_lines)
U, V = np.meshgrid(u, v)

# Convert grid to 3D Cartesian coordinates
X = np.sin(V) * np.cos(U)
Y = np.sin(V) * np.sin(U)
Z = np.cos(V)

# Convert grid to Latitude/Longitude
lon = np.degrees(U) - 180
lat = 90 - np.degrees(V)

# 2. Create the first plot: 3D Sphere
fig1 = plt.figure(figsize=(4, 3))
ax1 = fig1.add_subplot(111, projection='3d')
ax1.plot_wireframe(X, Y, Z, color='blue', alpha=0.5, linewidth=1)
ax1.set_title("Latitude and Longitude in 3D")
ax1.set_xlabel("$x$")
ax1.set_ylabel("$y$")
ax1.set_zlabel("$z$")
plt.show()

# 3. Create the second plot: 2D Projection
fig2 = plt.figure(figsize=(4, 3))
ax2 = fig2.add_subplot(111)
# Plot horizontal lines (constant latitude)
for i in range(len(lat)):
    ax2.plot(lon[i, :], lat[i, :], color='blue', alpha=0.3)
# Plot vertical lines (constant longitude)
for j in range(len(lon[0])):
    ax2.plot(lon[:, j], lat[:, j], color='blue', alpha=0.3)

ax2.set_title("Latitude and Longitude in 2D")
ax2.set_xlabel("Longitude (Degrees)")
ax2.set_ylabel("Latitude (Degrees)")
ax2.grid(True, linestyle='--', alpha=0.2)

plt.tight_layout()
plt.show()
```
