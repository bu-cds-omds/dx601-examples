# Video: Illustrating the Central Limit Theorem

This video illustrates the central limit theorem by showing how different sampled distributions look progressively more normal as the number of samples increases.


Script:
* Let's take a look at how the distribution of sample averages changes as predicted by the central limit theorem.
* I will start with an easy case where the samples are picked from the uniform distribution between zero and one.
* This should be easy because there is a limit range of values, and thus a limited variance, so the worst cases should not be aggressively triggered.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_mean_of_uniform(sample_size):
    samples = np.random.uniform(size=(100000, sample_size)).mean(axis=1)
    plt.hist(samples, bins=100)
```

Script:
* I will plot with one sample at a time first which should look like the original uniform distribution.

```python
plot_mean_of_uniform(1)
```

Script:
* That looks fairly flat as expected.
* What about two samples?

```python
plot_mean_of_uniform(2)
```

* This triangle pattern can be predicted analytically.
* The distribution really does have this triangle shape.

```python
plot_mean_of_uniform(3)
```

* With n=3 samples, now this is starting to look rounded like a normal distribution.

```python
plot_mean_of_uniform(4)
```

Script:
* The shape is changing slowly now.

```python
plot_mean_of_uniform(5)
```

Script:
* I am going to switch to n as powers of 10 now.

```python
plot_mean_of_uniform(10)
```

Script:
* Watch the x-axis.
* It is slowly getting more narrow as the distribution gets tighter.

```python
plot_mean_of_uniform(100)
```

```python
plot_mean_of_uniform(1000)
```

Script:
* The shape has been pretty consistent for a while, but the width has slowly been getting smaller.
* Let's try a harder distribution.
* I will make it Bernoulli with probability 1%.
* That will be a 1% chance of a 1, and a 99% chance of a zero.

```python
def plot_mean_of_bernoulli(sample_size):
    samples = (np.random.uniform(size=(100000, sample_size)) < 0.01).mean(axis=1)
    plt.hist(samples, bins=100)
```

Script:
* Like before, I will double check the distribution with n=1, one sample per average.

```python
plot_mean_of_bernoulli(1)
```

Script:
* Let's try n=2 now.

```python
plot_mean_of_bernoulli(2)
```

Script:
* Why are there only two bars?
* Each bar represents a different number of ones coming up in the Bernoulli sampling.
* Since the probability was set at 1%, it is very unlikely to have more than one sample.
* I'll increase the number samples faster this time.

```python
plot_mean_of_bernoulli(5)
```

```python
plot_mean_of_bernoulli(10)
```

```python
plot_mean_of_bernoulli(100)
```

Script:
* Finally, we are seeing more of a distribution, but it does not look normal yet.
* Pretty concentrated on the left, close to zero.

```python
plot_mean_of_bernoulli(1000)
```

Script:
* At n=1000, so one thousand samples averaged together for each point going into the histogram, we are seeing more of a normal distribution.
* At least a striped version of a normal distribution.
* Since we are starting with a discrete distribution, and in particular since it is zero and one values, every number going into this histogram is a multiple of one over n.
* In this case, that is a multiple of one thousandth.
* Zero point oh oh one.

```python
plot_mean_of_bernoulli(10000)
```

Script:
* Even at n = ten thousand, we see some artifacts of the discrete values, but it is starting to look more and more normal.
* And filled in if you don't squint too much or use too many histogram buckets.
* With both starting distributions, uniform random between zero and one, and Bernoulli with probability 1%, we eventually saw a distribution looking normal as we increased the sample sizes.
* At least, these distributions looked normal with the histograms.
* Averaging the uniform random samples converged to a normal distribution pretty quickly - it started looking normal at n=3 and kept improving from there.
* Averaging the Bernoulli samples took a lot longer; those did not look normal until around n=1000.
* I used to work in online marketing, and in a lot of scenarios, the ad response rates are around 1%.
* Some are even lower.
* So the second set of histograms matches a lot of real cases.
* If you have a rough idea of response rates, you can run some quick simulations like this to get a quick idea of the rate of convergence, and plan how long you will have to run a campaign or collect data.
* As you saw here, the central limit theorem does apply, but the speed to convergence varies widely.
