# Video: Visualizing Linear Dependence

This video shows how to visualize linear dependencies of low-dimensional vectors.

```python
import matplotlib.pyplot as plt
import numpy as np
```

Slide: Mental Model to Visualize Linear Dependence

* Treat the target vector as a position.
* Treat the other vectors as magnitudes and directions.
* Can the other vectors be combined to point at the target?


## A 2D Linear Dependence Problem


```python
plt.figure(figsize=(9,3))

plt.subplot(1, 3, 1)

plt.title("Initial Setup")

plt.gca().axhline(y=0, color="grey")
plt.gca().axvline(x=0, color="grey")
plt.gca().set_xlim((-2, 6))
plt.gca().set_ylim((-2, 6))

plt.arrow(0, 0, 3, 1, head_width=0.2, length_includes_head=True)
plt.arrow(0, 0, 1, 2, head_width=0.2, length_includes_head=True)

plt.plot(2, 1.5, marker="o")

plt.subplots_adjust(wspace=0.4)

None
```

Script:
* Consider the setup in this chart.
* Is the vector represented by the position marked by the blue dot linearly dependent on the two vectors drawn as arrows?


## Lining Up the 2D Linear Dependency


```python
plt.figure(figsize=(9,3))

plt.subplot(1, 3, 1)

plt.title("Initial Setup")

plt.gca().axhline(y=0, color="grey")
plt.gca().axvline(x=0, color="grey")
plt.gca().set_xlim((-2, 6))
plt.gca().set_ylim((-2, 6))

plt.arrow(0, 0, 3, 1, head_width=0.2, length_includes_head=True)
plt.arrow(0, 0, 1, 2, head_width=0.2, length_includes_head=True)

plt.plot(2, 1.5, marker="o")

plt.subplot(1, 3, 2)

plt.title("Slide Second Vector...")

plt.gca().axhline(y=0, color="grey")
plt.gca().axvline(x=0, color="grey")
plt.gca().set_xlim((-2, 6))
plt.gca().set_ylim((-2, 6))

plt.arrow(0, 0, 3, 1, head_width=0.2, length_includes_head=True)
plt.arrow(1.5, 0.5, 1, 2, head_width=0.2, length_includes_head=True)

plt.plot(2, 1.5, marker="o")

plt.subplots_adjust(wspace=0.4)

None
```

Script:
* We can visualize solving this by imagining one of the vectors sliding along the other vector until it points at, or overlaps the target position.

## Finishing the 2D Linear Dependency


```python
plt.figure(figsize=(9,3))

plt.subplot(1, 3, 1)

plt.title("Initial Setup")

plt.gca().axhline(y=0, color="grey")
plt.gca().axvline(x=0, color="grey")
plt.gca().set_xlim((-2, 6))
plt.gca().set_ylim((-2, 6))

plt.arrow(0, 0, 3, 1, head_width=0.2, length_includes_head=True)
plt.arrow(0, 0, 1, 2, head_width=0.2, length_includes_head=True)

plt.plot(2, 1.5, marker="o")

plt.subplot(1, 3, 2)

plt.title("Slide Second Vector...")

plt.gca().axhline(y=0, color="grey")
plt.gca().axvline(x=0, color="grey")
plt.gca().set_xlim((-2, 6))
plt.gca().set_ylim((-2, 6))

plt.arrow(0, 0, 3, 1, head_width=0.2, length_includes_head=True)
plt.arrow(1.5, 0.5, 1, 2, head_width=0.2, length_includes_head=True)

plt.plot(2, 1.5, marker="o")

plt.subplots_adjust(wspace=0.4)

plt.subplot(1, 3, 3)

plt.title("Rescale Magnitudes")

plt.gca().axhline(y=0, color="grey")
plt.gca().axvline(x=0, color="grey")
plt.gca().set_xlim((-2, 6))
plt.gca().set_ylim((-2, 6))

plt.arrow(0, 0, 1.5, 0.5, head_width=0.2, length_includes_head=True)
plt.arrow(1.5, 0.5, 0.5, 1.0, head_width=0.2, length_includes_head=True)

plt.plot(2, 1.5, marker="o")

plt.subplots_adjust(wspace=0.4)

None
```

Script:
* Then we rescale the vectors so they end at the next ones starting position or the target.
* The amount that those vectors are rescaled will be the coefficients in the linear dependency.

## Coverage From One Vector


```python
plt.figure(figsize=(9,3))

plt.subplot(1, 3, 1)

plt.title("Initial Setup")

plt.gca().axhline(y=0, color="grey")
plt.gca().axvline(x=0, color="grey")
plt.gca().set_xlim((-2, 6))
plt.gca().set_ylim((-2, 6))

plt.arrow(0, 0, 3, 1, head_width=0.2, length_includes_head=True)
plt.arrow(0, 0, 1, 2, head_width=0.2, length_includes_head=True)

plt.plot(2, 1.5, marker="o")

plt.subplot(1, 3, 2)

plt.title("Just One Vector Covers a Line")

plt.gca().axhline(y=0, color="grey")
plt.gca().axvline(x=0, color="grey")
plt.gca().set_xlim((-2, 6))
plt.gca().set_ylim((-2, 6))

for r in np.linspace(-2, 6):
    plt.arrow(0, 0, 3 * r, r, head_width=0.2, length_includes_head=True)

plt.plot(2, 1.5, marker="o")

plt.subplots_adjust(wspace=0.4)

None
```

Script:
* Just looking at the linear combinations from one vector, that gives us all the points on the line containing the origin and the vector as a position.

## Coverage From Two Independent Vectors


```python
plt.figure(figsize=(9,3))

plt.subplot(1, 3, 1)

plt.title("Initial Setup")

plt.gca().axhline(y=0, color="grey")
plt.gca().axvline(x=0, color="grey")
plt.gca().set_xlim((-2, 6))
plt.gca().set_ylim((-2, 6))

plt.arrow(0, 0, 3, 1, head_width=0.2, length_includes_head=True)
plt.arrow(0, 0, 1, 2, head_width=0.2, length_includes_head=True)

plt.plot(2, 1.5, marker="o")

plt.subplot(1, 3, 2)

plt.title("Just One Vector")

plt.gca().axhline(y=0, color="grey")
plt.gca().axvline(x=0, color="grey")
plt.gca().set_xlim((-2, 6))
plt.gca().set_ylim((-2, 6))

for r in np.linspace(-2, 6):
    plt.arrow(0, 0, 3 * r, r, head_width=0.2, length_includes_head=True)

plt.plot(2, 1.5, marker="o")

plt.subplot(1, 3, 3)

plt.title("Two Vector Mesh")

plt.gca().axhline(y=0, color="grey")
plt.gca().axvline(x=0, color="grey")
plt.gca().set_xlim((-2, 6))
plt.gca().set_ylim((-2, 6))

for r in np.linspace(-2, 6):
    plt.arrow(0, 0, 3 * r, r, head_width=0.2, length_includes_head=True)

for r1 in np.linspace(-2, 6):
    for r2 in np.linspace(-5, 10):
        plt.arrow(3 * r1, r1, 1 * r2, 2 * r2, head_width=0.2, length_includes_head=True)

plt.plot(2, 1.5, marker="o")

plt.subplots_adjust(wspace=0.4)

None
```

Script:
* If we get a second independent vector, then we get a plane from all the linear combinations.
* This chart still shows white space between the lines because I only plotted integer multiples.
* If all linear combinations were plotted, the chart would all be black.
* Generally, if we are only working with 2 dimensions, then 2 independent vectors cover all points with their linear combinations.
* So if there are 3 or more vectors in 2 dimensions, then there must be a linear dependency.
