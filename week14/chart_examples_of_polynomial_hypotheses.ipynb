```python
import matplotlib.pyplot as plt
import numpy as np

# Set seed for reproducibility
np.random.seed(42)

# Define x range
x = np.linspace(-2, 2, 200)
degrees = [2, 3, 5, 10]

for d in degrees:
    plt.figure(figsize=(6, 4))
    plt.axhline(0, color='lightgrey', linewidth=0.5)
    plt.axvline(0, color='lightgrey', linewidth=0.5)

    # Generate and plot 4 different polynomial functions of degree d
    for i in range(4):
        # Random coefficients for each power up to d
        coeffs = np.random.uniform(-2, 2, d + 1)

        # Evaluate polynomial: c0 + c1*x + c2*x^2 + ...
        y = np.polyval(coeffs[::-1], x)

        plt.plot(x, y, label=f'Sample {i+1}')

    plt.title(f'Polynomial Hypotheses (Degree {d})')
    plt.xlabel('x')
    plt.ylabel('h(x)')
    plt.xlim(-2, 2)
    plt.ylim(-10, 10)
    plt.legend(fontsize='small')
    plt.grid(True, linestyle='--', alpha=0.3)
    plt.show()
```
