```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn.linear_model

xs = np.linspace(-5, 5, 100)
ys_target = xs**2 + xs

m = sklearn.linear_model.LinearRegression()
m.fit(xs.reshape(-1, 1), ys_target)
ys_predicted = m.predict(xs.reshape(-1, 1))

plt.plot(xs, ys_target, label="target")
plt.plot(xs, ys_predicted, label="predicted")

# Highlight approximation error
plt.fill_between(xs, ys_target, ys_predicted, color='red', alpha=0.2, label="approximation error")

plt.legend()
plt.xlabel("$x$")
plt.xlim(-5, 5)
plt.ylabel("$y$")

plt.show()
```
