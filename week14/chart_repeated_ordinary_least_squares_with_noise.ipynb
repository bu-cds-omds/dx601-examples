```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

np.random.seed(42_14_3)

xs = np.linspace(-5, 5, 20)
sigma=1

plt.figure(figsize=(8, 6))

plt.plot(xs, xs, label="ground truth (y=x)", color="black", linewidth=5)

for i in range(10):
    ys = xs + np.random.normal(scale=sigma, size=xs.shape)
    model = LinearRegression(fit_intercept=False)
    model.fit(xs.reshape(-1, 1), ys)
    plt.plot(xs, model.predict(xs.reshape(-1, 1)), alpha=0.5, label=f"model {i}")

plt.axhline(y=0, color="lightgray")
plt.axvline(x=0, color="lightgray")
plt.legend()
plt.title(f"Repeated Ordinary Least Squares with Noise ($\\sigma^2={sigma}^2$)")
plt.xlabel("$x$")
plt.xlim(-5, 5)
plt.ylabel("$y$")

plt.show()
```
