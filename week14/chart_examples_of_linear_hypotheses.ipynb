```python
import matplotlib.pyplot as plt
import numpy as np

# Set seed for reproducibility
np.random.seed(42)

# Define x range
x = np.linspace(-10, 10, 100)

plt.figure(figsize=(6, 4))

plt.axhline(0, color='lightgrey', linewidth=0.5)
plt.axvline(0, color='lightgrey', linewidth=0.5)

# Generate and plot 4 different linear functions
for i in range(4):
    # Random coefficients between -2 and 2
    theta = np.random.uniform(-2, 2)
    b = np.random.uniform(-2, 2)

    y = theta * x + b

    plt.plot(x, y, label=f'h(x) = {theta:.2f}x + {b:.2f}')

plt.xlabel('x')
plt.ylabel('h(x)')
plt.legend()
plt.show()
```
