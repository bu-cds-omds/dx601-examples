```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)

xs = np.linspace(-5, 5, 200)
ys_expected = xs**2
ys_target = ys_expected + np.random.normal(scale=5, size=xs.shape)

plt.plot(xs, ys_target, label="target samples", linestyle="none", marker=".")
plt.plot(xs, ys_expected, label="best possible")

# Highlight irreducible error
plt.fill_between(xs, ys_expected, ys_target, color='red', alpha=0.2, label="irreducible error", step="mid")

plt.legend()
plt.title("Random Samples vs Best Possible Predictions")
plt.xlabel("$x$")
plt.xlim(-5, 5)
plt.ylabel("$y$")

plt.show()
```
