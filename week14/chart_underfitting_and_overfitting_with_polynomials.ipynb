```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)

# Define ground truth function
def f(x):
    return x + np.sin(x)

# Generate ground truth line
xs_ground_truth = np.linspace(-5, 5, 200)
ys_ground_truth = f(xs_ground_truth)

# Generate 20 noisy samples
x_samples = np.random.uniform(-5, 5, 20)
y_samples = f(x_samples) + np.random.normal(scale=0.5, size=x_samples.shape)

# Plotting
plt.figure(figsize=(8, 6))
plt.plot(xs_ground_truth, ys_ground_truth, label="ground truth", color='black', linewidth=2, linestyle='--')
plt.scatter(x_samples, y_samples, color='red', label="samples", alpha=0.7)

plt.title("Ground Truth vs. Noisy Samples")
plt.xlabel("$x$")
plt.xlim(-5, 5)
plt.ylabel("$y$")
plt.ylim(-10, 10)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# --- Plot 2: Comparison of Polynomial Models ---
# Fit and plot polynomial degrees 1, 6, and 18
for degree, color, comment in zip([1, 6, 18], ['tab:blue', 'tab:green', 'tab:orange'], ["Underfitting", "Reasonable", "Overfitting"]):
    coeffs = np.polyfit(x_samples, y_samples, degree)
    model = np.poly1d(coeffs)

    plt.figure(figsize=(8, 6))

    # Plot ground truth and samples for context
    plt.plot(xs_ground_truth, ys_ground_truth, label="ground truth", color='black', linewidth=2, linestyle='--')
    plt.scatter(x_samples, y_samples, color='red', label="samples", alpha=0.6, edgecolors='k')

    plt.plot(xs_ground_truth, model(xs_ground_truth), label=f"Degree {degree} Fit", color=color, linewidth=2)

    plt.title(f"Degree {degree} Polynomial ({comment})")
    plt.xlabel("$x$")
    plt.xlim(-5, 5)
    plt.ylabel("$y$")
    plt.ylim(-10, 10) # Manage high degree oscillations
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()
```
