## Dependence of Two Random Rows

\begin{bmatrix}
v_{0,0} & v_{0,1} & v_{0,2} & v_{0,3} & v_{0,4} & v_{0,5} & v_{0,6} & v_{0,7} \\
v_{1,0} & \frac{v_{1,0}}{v_{0,0}}v_{0,1} & \frac{v_{1,0}}{v_{0,0}}v_{0,2} & \frac{v_{1,0}}{v_{0,0}}v_{0,3} & \frac{v_{1,0}}{v_{0,0}}v_{0,4} & \frac{v_{1,0}}{v_{0,0}}v_{0,5} & \frac{v_{1,0}}{v_{0,0}}v_{0,6} & \frac{v_{1,0}}{v_{0,0}}v_{0,7}\\
\end{bmatrix}

```python
import matplotlib.pyplot as plt
import numpy as np

xs = np.linspace(-2, 2)

sample_x = 0.54
sample_y = 0.76

plt.plot(xs, xs * (sample_y / sample_x))
plt.plot(0, 0, linestyle=None, marker="*", markersize=10, label="origin")
plt.plot(sample_x, sample_y, linestyle=None, marker="*", markersize=10, label="first data point")

plt.axhline(y=0, color="grey")
plt.axvline(x=0, color="grey")
plt.legend()
plt.gca().set_xlim(-2, 2)
plt.gca().set_ylim(-2, 2)
plt.title("Will the next data point fall on this line?")

None
```
