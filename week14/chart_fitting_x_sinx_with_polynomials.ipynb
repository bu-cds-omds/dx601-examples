```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import mean_squared_error

np.random.seed(42_14_4)

xs = np.linspace(-5, 5, 100)
ys = xs + np.cos(xs)

for d in range(1, 11):
    plt.figure(figsize=(4, 3))

    plt.plot(xs, ys, label="ground truth", linestyle='--')

    # Fit a polynomial of degree d
    coeffs = np.polyfit(xs, ys, d)
    poly = np.poly1d(coeffs)
    ys_pred = poly(xs)

    # Calculate MSE
    mse = mean_squared_error(ys, ys_pred)

    plt.plot(xs, ys_pred, label=f"degree {d} fit")

    plt.title(f"Degree {d} Polynomial (MSE: {mse:.4f})")
    plt.xlabel("$x$")
    plt.xlim(-5, 5)
    plt.ylabel("$y$")
    plt.legend()
    plt.tight_layout()
    plt.show()
```
